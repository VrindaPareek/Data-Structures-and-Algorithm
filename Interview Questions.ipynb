### Happy Number

```
Write an algorithm to determine if a number is "happy".

A happy number is a number defined by the following process: Starting with any positive integer, replace the number by the sum of the squares of its digits, and repeat the process until the number equals 1 (where it will stay), or it loops endlessly in a cycle which does not include 1. Those numbers for which this process ends in 1 are happy numbers.

Example 1:

Input: 19
Output: true
Explanation:

1^{2} + 9^{2} = 82
8^{2} + 2^{2} = 68
6^{2} + 8^{2} = 100
1^{2} + 0^{2} + 0^{2} = 1
```

### Solution
```python
def isHappy(n):

    def get_sum(n):
        num=[]
        while n>1:
            digit=n%10
            num.append(digit)
            n=n//10
        num.append(n)

        sum_num=0
        for d in num:
            sum_num+=d**2
        return sum_num  

    check=set()
    sum_num=get_sum(n)
    while sum_num!=1:
        sum_num=get_sum(sum_num)
        if sum_num in check:
            return False
        else:
            check.add(sum_num)
    return True

isHappy(19)
```

### Single Number

```
Given a non-empty array of integers, every element appears twice except for one. Find that single one.

Your algorithm should have a linear runtime complexity. Could you implement it without using extra memory?

Example 1:

Input: [2,2,1]
Output: 1

Example 2:

Input: [4,1,2,1,2]
Output: 4

```

### Solution
```python
arr=[3,3,5,6,5]
result=0
for each in arr:
    result=result^each # property A xor A = 0, 0 xor A = A, A xor B = B xor A
result
```

**Time complexity : O(n)**

**Space complexity : O(1)**

### Question (Amazon)

Given a sorted array arr of distinct integers, return the lowest index i for which arr[i] == i. Return null if there is no such index.

For example, given the array [-5, -3, 2, 3], return 2 since arr[2] == 2. Even though arr[3] == 3, we return 2 since it's the lowest index

### Solution

```python
def match(arr):
    low = 0
    high = len(arr)-1
    while (low <= high):
        middle = (low + high)//2
        if (arr[middle] == middle):
            return middle
        if (arr[middle] < middle):
            low = middle + 1
        else:
            high = middle - 1
    print('No match found')
    return None

match([-1,0,1,3,5,6,7])
```
**Complexity**: O(logn) due to binary search, n is the size of array

### Question (Microsoft)
You are given a list of jobs to be done, where each job is represented by a start time and end time. Two jobs are compatible if they don't overlap. Find the largest subset of compatible jobs. 

For example, given the following jobs (there is no guarantee that jobs will be sorted):
[(0, 6),
(1, 4),
(3, 5),
(3, 8),
(4, 7),
(5, 9),
(6, 10),
(8, 11)]

Output: [(1, 4),
(4, 7),
(8, 11)]


### Solution
```python
def compatible_jobs(jobs):
    jobs_in_finish_order=sorted(jobs,key=lambda x:x[1])
    non_overlapping_jobs=[]
    p=jobs_in_finish_order[0]
    non_overlapping_jobs.append(p)
    for i in range(len(jobs_in_finish_order)-1):
        if p[1]<=jobs_in_finish_order[i+1][0]:
            non_overlapping_jobs.append(jobs_in_finish_order[i+1])
            p=jobs_in_finish_order[i+1]
    return(non_overlapping_jobs)
    
compatible_jobs([(0, 6), (1, 4), (3, 5), (3, 8), (4, 7), (5, 9), (6, 10), (8, 11)])
```
**Complexity**: O(nlogn) where n is the size of array

### Question (Robinhood)
Given an array of strings, group anagrams together.

**Input**: ["eat", "tea", "tan", "ate", "nat", "bat"],

**Output**:
[
  ["ate","eat","tea"],
  ["nat","tan"],
  ["bat"]
]

### Solution  

```python
import collections
def groupAnagrams(strs):
    ans = collections.defaultdict(list)
    for s in strs:
        ans[tuple(sorted(s))].append(s)
    return ans.values()

groupAnagrams(["eat", "tea", "tan", "ate", "nat", "bat"])
```

**Complexity**: O(nklogk) where n is the size of array and k is the length of maximum substring

### Question (Amazon)

Write a function that takes a natural number as input and returns the number of digits the input has.

Constraint: don't use any loops.

### Solution

```python
import math  
def countDigit(n): 
    return math.floor(math.log(n, 10)+1) 
  
n = int(input())
print ("Number of digits : % d"%(countDigit(n))) 
```

### Question (Facebook)

Given a list of strictly positive integers, partition the list into 3 contiguous partitions which each sum up to the same value. If not possible, return null.

For example, given the following list: [3, 5, 8, 0, 8]

Return the following 3 partitions:

[[3, 5],
 [8, 0],
 [8]]

which each add up to 8.

### Answer
```python
def partition(arr):
    arr_sum=sum(arr)
    if arr_sum%3!=0:
        print("Not Possible")
        return None
    else:
        first_sum=arr_sum//3
        second_sum=2*first_sum
        
        detect_sum1=0
        first_partition=[]
        pos=0
        for each in arr:
            first_partition.append(each)
            detect_sum1=detect_sum1+each
            if detect_sum1==first_sum:
                pos+=1
                break
            pos+=1
            
        second_partition=[]
        detect_sum2=0
        detect_sum3=detect_sum1
        for each in arr[pos:]:
            second_partition.append(each)
            detect_sum2=detect_sum2+each
            detect_sum3=detect_sum3+each
            if (detect_sum2==first_sum) & (detect_sum3%second_sum==0):
                pos+=1
                break
            pos+=1
        third_partition=arr[pos:]
        if sum(third_partition)==first_sum:
            return [first_partition,second_partition,third_partition]
        else:
            return None
        
arr=[3,5,8,0,8]
partition(arr)
```
**Complexity:** O(n) where n is the length of the array




### Question (Twitter)

A strobogrammatic number is a positive number that appears the same after being rotated 180 degrees. For example, 16891 is strobogrammatic.

Create a program that finds all strobogrammatic numbers with N digits.

### Solution

```python
def is_strobogrammatic(number):
    reverse=''
    for i in range(len(number)-1,-1,-1):
        if number[i] in {'1','0','8'}:
            reverse=reverse+number[i]
        elif number[i]=='6':
            reverse=reverse+'9'
        elif number[i]=='9':
            reverse=reverse+'6'       
        else:
            return False
    if reverse == number:
        return True
    else:
        return False
    
n=int(input("Enter N: "))
result=[]
for i in range((10**(n-1)), (10**n)):
    if is_strobogrammatic(str(i)):
        result.append(i)
print(result)
```
