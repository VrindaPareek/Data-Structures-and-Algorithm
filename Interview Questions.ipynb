### Question
Given an array of strings, group anagrams together.

**Input**: ["eat", "tea", "tan", "ate", "nat", "bat"],

**Output**:
[
  ["ate","eat","tea"],
  ["nat","tan"],
  ["bat"]
]

### Solution  

```python
import collections
def groupAnagrams(strs):
    ans = collections.defaultdict(list)
    for s in strs:
        ans[tuple(sorted(s))].append(s)
    return ans.values()

groupAnagrams(["eat", "tea", "tan", "ate", "nat", "bat"])
```

**Complexity**: O(nklogk) where n is the size of array and k is the length of maximum substring

You are given a list of jobs to be done, where each job is represented by a start time and end time. Two jobs are compatible if they don't overlap. Find the largest subset of compatible jobs. 

For example, given the following jobs (there is no guarantee that jobs will be sorted):
[(0, 6),
(1, 4),
(3, 5),
(3, 8),
(4, 7),
(5, 9),
(6, 10),
(8, 11)]

Output: [(1, 4),
(4, 7),
(8, 11)]


### Solution
```python
def compatible_jobs(jobs):
    jobs_in_finish_order=sorted(jobs,key=lambda x:x[1])
    non_overlapping_jobs=[]
    p=jobs_in_finish_order[0]
    non_overlapping_jobs.append(p)
    for i in range(len(jobs_in_finish_order)-1):
        if p[1]<=jobs_in_finish_order[i+1][0]:
            non_overlapping_jobs.append(jobs_in_finish_order[i+1])
            p=jobs_in_finish_order[i+1]
    return(non_overlapping_jobs)
    
compatible_jobs([(0, 6), (1, 4), (3, 5), (3, 8), (4, 7), (5, 9), (6, 10), (8, 11)])
```
**Complexity**: O(nlogn) where n is the size of array

### Question

Given a sorted array arr of distinct integers, return the lowest index i for which arr[i] == i. Return null if there is no such index.

### Solution

```python
def match(arr):
    low = 0
    high = len(arr)-1
    while (low <= high):
        middle = (low + high)//2
        if (arr[middle] == middle):
            return middle
        if (arr[middle] < middle):
            low = middle + 1
        else:
            high = middle - 1
    print('No match found')
    return None

match([-1,0,1,3,5,6,7])
```

### Question

Write a function that takes a natural number as input and returns the number of digits the input has.

Constraint: don't use any loops.

### Solution

```python
import math  
def countDigit(n): 
    return math.floor(math.log(n, 10)+1) 
  
n = int(input())
print ("Number of digits : % d"%(countDigit(n))) 
```
